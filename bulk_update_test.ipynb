```python
def clean_name(name):
    """Clean name by removing parentheses and normalizing"""
    if not name:
        return ""

    # Remove parentheses and their contents
    cleaned = name.split('(')[0]

    # Remove special characters
    cleaned = ''.join(char for char in cleaned if char.isalnum() or char.isspace())

    # Normalize whitespace and convert to lowercase
    return ' '.join(cleaned.split()).lower().strip()
```

```python
#deprecated version
from ruamel.yaml import YAML
from ruamel.yaml.comments import CommentedMap

class OrderedYAML(YAML):
    def __init__(self):
        super().__init__()
        self.preserve_quotes = True
        # self.indent(sequence=2)
        # self.width = 4096
        # self.default_flow_style = False
        # self.map_format = 'rt'
def update_yaml_with_collabs(yaml_file_path, dataset_name, category):
    """Update YAML file using both exact and cleaned name matching"""
    try:
        with open(yaml_file_path, 'r') as file:
            lines = file.readlines()  # Read all lines while preserving format

            # Load YAML for name comparison
            yaml = OrderedYAML()
            data = yaml.load(''.join(lines))

            yaml_name = data.get('Name', '')

            # Try exact match first, then cleaned match
            if yaml_name == dataset_name or clean_name(yaml_name) == clean_name(dataset_name):
                if 'DeprecatedNotice' in data and isinstance(data['DeprecatedNotice'], str) and data['DeprecatedNotice'].strip():
                    return 'deprecated'

                print(f"\nMatch found!")
                print(f"Excel name: {dataset_name}")
                print(f"YAML name: {yaml_name}")

                try:
                    updated_lines = []
                    modified = False

                    # Check if Collabs already exists
                    has_collabs = 'Collabs:' in ''.join(lines)

                    # Process lines
                    i = 0
                    while i < len(lines):
                        line = lines[i]

                        # If we find Tags and Collabs doesn't exist, add Collabs before Tags
                        if line.strip().startswith('Tags:') and not has_collabs:
                            # Add Collabs section first
                            updated_lines.append('Collabs:\n')
                            updated_lines.append('  ASDI:\n')
                            updated_lines.append(f'    Tags:\n')
                            updated_lines.append(f'      - {category.strip()}\n')
                            modified = True

                            # Then add Tags section
                            updated_lines.append(line)
                        else:
                            updated_lines.append(line)
                        i += 1

                    # Only write if changes were made
                    if modified:
                        with open(yaml_file_path, 'w') as outfile:
                            outfile.writelines(updated_lines)
                        print(f"Successfully updated {yaml_file_path}")
                        return 'modified'
                    else:
                        print(f"No changes needed for {yaml_file_path}")
                        return 'matched'

                except Exception as write_error:
                    print(f"Error writing file: {write_error}")
                    raise

    except Exception as e:
        print(f"Error processing {yaml_file_path}: {e}")

    return False


def process_yaml_files(dataset_folder, excel_file):
    """Process all YAML files and match with Excel data"""
    total_yaml_files = 0
    processed_files = 0
    matches_found = 0
    matched_dataset_names = []
    failed_files = []
    git_files = []  # Track YAML files to be updated
    deprecated_matched = []  # Track deprecated but matched datasets
    # Read Excel file
    try:
        df = pd.read_excel(excel_file)
        # Ensure the required columns exist
        if 'Dataset' not in df.columns or 'Category' not in df.columns:
            raise ValueError("Excel file must contain 'Name' and 'Category' columns")

        # Create dictionary of dataset names and categories
        dataset_dict = dict(zip(df['Dataset'], df['Category']))
        # print(dataset_dict)

        yaml_folder_path = Path(dataset_folder)
        if not yaml_folder_path.exists() or not yaml_folder_path.is_dir():
            raise ValueError(f"Invalid dataset folder path: {dataset_folder}")

        # Count total YAML files
        total_yaml_files = len([f for f in yaml_folder_path.glob('*.yaml')])
        print("\n=== Matching starts ===")
        for yaml_file in yaml_folder_path.glob('*.yaml'):
            processed_files += 1

            try:
                for dataset_name, category in dataset_dict.items():
                    match_result = update_yaml_with_collabs(yaml_file, dataset_name, category)
                    if match_result == 'deprecated':
                        deprecated_matched.append({
                            'dataset': dataset_name,
                            'file': yaml_file.name
                        })
                        matched_dataset_names.append(dataset_name)
                        print(f"Dataset {dataset_name} matched but marked as deprecated - skipping update")
                        break
                    elif match_result == 'modified':
                        matches_found += 1
                        matched_dataset_names.append(dataset_name)
                        git_files.append(yaml_file.name)
                        break
                    elif match_result == 'matched':
                        matches_found += 1
                        matched_dataset_names.append(dataset_name)
                        break



            except Exception as e:
                print(f"Error processing {yaml_file}: {e}")
                failed_files.append(str(yaml_file))


        # Final unmatched datasets
        final_unmatched = df[~df['Dataset'].isin(matched_dataset_names)]

        # Save unmatched datasets to Excel
        output_path = 'unmatched_datasets.xlsx'
        final_unmatched.to_excel(output_path, index=False)

        # Print final summary
        print("\n=== Final Processing Summary ===")
        print(f"Total YAML files: {total_yaml_files}")
        print(f"Total matches found: {matches_found}")
        print(f"Failed files: {len(failed_files)}")
        print(f"Final unmatched datasets: {len(final_unmatched)}")
        print(f"Deprecated matches (not updated): {len(deprecated_matched)}")
        # print(f"Unmatched datasets saved to: {output_path}")
        print(f"\nTotal Git files to be updated: {len(git_files)}")
        print("\nDeprecated matched datasets:")
        for dep in deprecated_matched:
                print(f"Dataset: {dep['dataset']} - File: {dep['file']}")
        return {
        'total_files': total_yaml_files,
        'total_matches': matches_found,
        'failed_files': failed_files,
        'matched_datasets': matched_dataset_names,
        'unmatched_count': len(final_unmatched),
        'deprecated_matched': deprecated_matched,
        'git_files': git_files
            }

    except Exception as e:
        print(f"Error in main processing: {e}")
        return None

```
